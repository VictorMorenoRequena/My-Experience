```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
diamonds= pd.read_csv("/Users/victormorenorequena/Desktop/Python_Youtube/diamonds.csv")
diamonds=diamonds.drop("Unnamed: 0", axis=1)
print(diamonds.shape)
diamonds.head(5)
```

```python
diamonds.hist(column="carat",
              figsize=(8,8),
              color="blue")
```

```python
diamonds.hist(column="carat",
              figsize=(8,8),
              color="blue",
              bins=50,
              range=(0,3.5))
```

```python
#Boxplots
```

```python
diamonds.boxplot(column="carat")
```

```python
diamonds.boxplot(column="price",
                 by= "clarity",
                 color="blue",
                 figsize=(8,8))
```

```python
#Density plots
```

```python
diamonds["carat"].plot(kind="density",
                       figsize=(8,8),
                       xlim=(0,5))
```

```python
carat_table= pd.crosstab(index=diamonds["clarity"], columns="count")
carat_table
```

```python
carat_table.plot(kind="bar",
                 figsize=(8,8))
```

```python
carat_table= pd.crosstab(index=diamonds["clarity"], 
                         columns=diamonds["color"])
carat_table
```

```python
carat_table.plot(kind="bar",
                figsize=(8,8),
                stacked= True)
```

```python
carat_table.plot(kind="bar",
                figsize=(8,8),
                stacked= False)
```

```python
#Scatterplotts
```

```python
diamonds.plot(kind="scatter",
              x= "carat",
              y="price",
              figsize=(10,10),
              ylim=(0,20000))
```

```python
#line plot
```

```python
years = [y for y in range(1956,2016)]

readings = [(y+np.random.uniform(0,20)-1900)for y in years]

time_df = pd.DataFrame({"year": years,
                        "readings": readings})

time_df.plot(x="year",
             y="readings",
             figsize=(9,9))
```

```python
#Saving image plot on pandas
```

```python
my_plot = time_df.plot(x="year",
             y="readings",
             figsize=(9,9))

my_fig = my_plot.get_figure()

my_fig.savefig("line_plot_example.png")
```
