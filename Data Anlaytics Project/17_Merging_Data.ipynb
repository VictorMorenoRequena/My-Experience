```python
import numpy as np
import pandas as pd
import os
```

```python
table_1 = pd.DataFrame({"P_ID": (1,2,4,5,7,8,9,10),
                        "gender":("male", "male", "female", "female",
                                  "female", "male", "male", "female"),
                        "height": (71,73,64,64,66,69,62,72),
                         "weight": (175,225,130,125,165,160,115,250)})
table_1
```

```python
table_2= pd.DataFrame({"P_ID": (1,2,3,4,5,6,7,8),
                       "sex":("male", "male", "female", "female", "female", "male", "male", "female"),
                       "visits": (1,2,4,12,2,2,1,1),
                       "checkups": (1,1,1,1,1,1,0,0),
                       "follow_up": (0,0,1,2,0,0,0,0),
                       "illness": (0,0,2,7,1,1,0,0),
                       "surgery": (0,0,0,2,0,0,0,0),
                       "ER": (0,1,0,0,0,0,1,1)})
table_2
```

```python
combined1 = pd.merge(table_1, table_2, how="inner", on="P_ID")
combined1
```

```python
left_join = pd.merge(table_1, table_2, how="left", on="P_ID")
left_join
```

```python
right_join = pd.merge(table_1, table_2, how="right", on="P_ID")
right_join
```

```python
outer_join = pd.merge(table_1, table_2, how="outer", on="P_ID")
outer_join
```

```python
table_1.rename(columns={"gender":"sex"}, inplace=True)
combined2 = pd.merge(table_1,table_2, how="outer",on=["P_ID", "sex"])
combined2
```
